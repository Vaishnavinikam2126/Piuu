```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('C:\\Users\\vaish\Downloads\\insurance_data.csv')
df
```

```python
%matplotlib inline
plt.xlabel('age')
plt.ylabel('bought_insurance')
plt.scatter(df.age,df.bought_insurance,color='Green',marker='*')
plt.show()
```

```python
new_df=df.drop('bought_insurance',axis='columns')
new_df
```

```python
bought_insurance=df.bought_insurance
bought_insurance
```

```python
# create Linear regression object
reg=linear_model.LinearRegression()
```

```python
reg.fit(new_df,bought_insurance)
```

## 1. Predict price of a home with Area=4000 sqr ft

```python
reg.predict([[4000]])
```
